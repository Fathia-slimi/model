```python
from flask import Flask, render_template, request, jsonify
from pymongo import MongoClient
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

app = Flask(__name__)

# Connexion à la base de données MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["nom_de_la_base_de_données"]
collection = db["nom_de_la_collection"]

@app.route("/classify", methods=["POST"])
def classify():
    # Récupérer les positions depuis la base de données MongoDB
    positions = []
    for doc in collection.find():
        positions.append(doc["positions"])

    # Convertir les positions en un dataframe pandas
    positions_df = pd.DataFrame(positions)

    # Remplacer les valeurs manquantes par la médiane de chaque colonne
    positions_df['x'].fillna(positions_df['x'].median(), inplace=True)
    positions_df['y'].fillna(positions_df['y'].median(), inplace=True)

    # Récupérer les valeurs de x et y
    coord_x = positions_df['x'].values
    coord_y = positions_df['y'].values

    # Transformer les points en matrice numpy
    coordinates = np.column_stack((coord_x, coord_y))

    # Normaliser les données
    scaler = StandardScaler()
    coordinates = scaler.fit_transform(coordinates)

    # Entraîner le modèle DBSCAN sur les points
    dbscan = DBSCAN(eps=0.5, min_samples=1)  # eps est défini à 0.5 comme exemple, vous pouvez ajuster ce paramètre selon votre besoin
    model = dbscan.fit(coordinates)

    # Prédire à quel cluster appartient chaque point
    labels = model.labels_

    # Compter le nombre de clusters formés
    num_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    print(num_clusters)

    # Classifier l'image en dyslexique ou non dyslexique en fonction du nombre de clusters formés
    if num_clusters > 10:
        result = "dyslexique"
    else:
        result = "non dyslexique"

    # Retourner la prédiction sous forme de JSON
    return jsonify(result=result)

if __name__ == "__main__":
    app.run(debug=True)

```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score

# Initialiser la liste des coordonnées pour chaque image
all_coordinates = []
# Initialiser la liste des étiquettes de dyslexie pour chaque image
all_labels = []

# Boucle sur toutes les images
for i in range(len(data.positions)):
    positions = pd.DataFrame(data.positions[i])
    
    # Remplacer les valeurs manquantes par la médiane de chaque colonne
    positions['x'].fillna(positions['x'].median(), inplace=True)
    positions['y'].fillna(positions['y'].median(), inplace=True)
    # Récupérer les valeurs de x et y
    coord_x = positions['x'].values
    coord_y = positions['y'].values
    # Transformer les points en matrice numpy
    coordinates = np.column_stack((coord_x, coord_y))
    all_coordinates.append(coordinates)
    is_dyslexic = data.dyslexic[i]
    is_dyslexic = 1 if is_dyslexic == 'Yes' else 0
    all_labels.append(np.full(len(coordinates), is_dyslexic))
    
# Vérifier que all_labels n'est pas vide
if len(all_labels) == 0:
    raise ValueError("all_labels is empty")

# Vérifier que chaque élément de all_labels est un tableau de dimension supérieure à zéro
for label_array in all_labels:
    if label_array.ndim == 0:
        raise ValueError("Element of all_labels is a zero-dimensional array")

# Concaténer toutes les coordonnées en une seule matrice
coordinates = np.concatenate(all_coordinates, axis=0)

# Concaténer toutes les étiquettes en une seule liste
labels = np.concatenate(all_labels, axis=0)

# Séparer les données en un ensemble d'entraînement et un ensemble de test
X_train, X_test, y_train, y_test = train_test_split(coordinates , labels, test_size=0.2, random_state=42)

# Normaliser les données d'entraînement
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)

# Entraîner le modèle DBSCAN sur les données d'entraînement
dbscan = DBSCAN(eps=0.3398911994502297, min_samples=2, metric='euclidean')
model = dbscan.fit(X_train)

# Prédire à quel cluster appartient chaque point dans les données de test
X_test = scaler.transform(X_test)
labels = model.fit_predict(X_test)

# Évaluer les performances du modèle en comparant les prédictions aux véritables étiquettes
accuracy = accuracy_score(y_test, labels)
confusion_mat = confusion_matrix(y_test, labels)
print("Accuracy:", accuracy)
print("Confusion matrix:", confusion_mat)

```

```python
import itertools
import matplotlib.pyplot as plt
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
  
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j]),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
from keras.applications import VGG16
from keras.layers import Dense, Dropout, Flatten
from keras.models import Sequential
from keras.optimizers import Adam
from keras.preprocessing.image import ImageDataGenerator
img_width, img_height = 224, 224
from keras import backend as K
from tensorflow.keras.models import Model
```

```python
if K.image_data_format() == 'channels_first':
    input_shape = (3, img_width, img_height)
else:
    input_shape = (img_width, img_height, 3)
```

```python
# Charger le modèle VGG16 pré-entraîné sans les couches fully-connected
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(img_width, img_height, 3))

for layer in base_model.layers:
    layer.trainable = False

x = base_model.output
x = Flatten()(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(1, activation='sigmoid')(x)

# Créer un nouveau modèle à partir des couches du modèle de base et des couches fully-connected personnalisées
model = Model(inputs=base_model.input, outputs=predictions)



```

```python
from keras.callbacks import ModelCheckpoint
checkpoint = ModelCheckpoint('model-{epoch:03d}-{accuracy:03f}-{val_accuracy:03f}.h5', 
                              verbose=1, 
                              monitor='val_loss',
                              save_best_only=True, 
                              mode='auto')
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
train_datagen = ImageDataGenerator(
	rescale=1. / 255,
	horizontal_flip=False)

test_datagen = ImageDataGenerator(rescale=1. / 255)

train_generator = train_datagen.flow_from_directory(
	train_data_dir,
	target_size=(img_width, img_height),
	batch_size=batch_size,
	seed=42,
	shuffle=True,
	class_mode='binary')

test_generator = test_datagen.flow_from_directory(
    test,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False,
    seed=42)
validation_generator = test_datagen.flow_from_directory(
	validation_data_dir,
	target_size=(img_width, img_height),
	batch_size=batch_size,
	seed=42,
	shuffle=False,
	class_mode='binary')

```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
earlystop = EarlyStopping(monitor='val_loss', mode='min', patience=4)
```

```python

history= model.fit(
	train_generator,
	steps_per_epoch=nb_train_samples // batch_size,
	epochs=epochs,
	callbacks=[checkpoint],
	verbose=False,
	validation_data=validation_generator,
	validation_steps=nb_validation_samples // batch_size)
```

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
  
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j]),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
# Importing all necessary libraries
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras import backend as K
  
img_width, img_height = 224, 224
```

```python
if K.image_data_format() == 'channels_first':
    input_shape = (3, img_width, img_height)
else:
    input_shape = (img_width, img_height, 3)
```

```python
model1 = Sequential()
model1.add(Conv2D(32, (2, 2), input_shape=input_shape))
model1.add(Activation('relu'))
model1.add(MaxPooling2D(pool_size=(2, 2)))
  
model1.add(Conv2D(32, (2, 2)))
model1.add(Activation('relu'))
model1.add(MaxPooling2D(pool_size=(2, 2)))
  
model1.add(Conv2D(64, (2, 2)))
model1.add(Activation('relu'))
model1.add(MaxPooling2D(pool_size=(2, 2)))
  
model1.add(Flatten())
model1.add(Dense(64))
model1.add(Activation('relu'))
model1.add(Dropout(0.5))
model1.add(Dense(1))
model1.add(Activation('sigmoid'))
```

```python
model1.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
train_datagen = ImageDataGenerator(
	rescale=1. / 255,
	
	horizontal_flip=False)

test_datagen = ImageDataGenerator(rescale=1. / 255)

train_generator = train_datagen.flow_from_directory(
	train_data_dir,
	target_size=(img_width, img_height),
	batch_size=batch_size,
	class_mode='binary')

test = test_datagen.flow_from_directory(
        test,
        target_size=(224, 224),
        batch_size=10,
        )

validation_generator = test_datagen.flow_from_directory(
	validation_data_dir,
	target_size=(img_width, img_height),
	batch_size=batch_size,
	class_mode='binary')

history= model1.fit(
	train_generator,
	steps_per_epoch=nb_train_samples // batch_size,
	epochs=epochs,
	callbacks=[checkpoint],
	verbose=False,
	validation_data=validation_generator,
	validation_steps=nb_validation_samples // batch_size)


```
